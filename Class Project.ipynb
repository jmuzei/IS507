## 2. Preprocessing, feature extraction and feature selection

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, :-1].values
y = df['Rating'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel


# Create a multinomial logistic regression object
mlr = LogisticRegression(multi_class='multinomial', solver='lbfgs')

# Fit the multinomial logistic regression model to the training data
mlr.fit(X_train, y_train)

# Use SelectFromModel to select the most important features based on the multinomial logistic regression coefficients
sfm = SelectFromModel(mlr, threshold=0.5, max_features=10)
X_train_selected = sfm.fit_transform(X_train, y_train)
X_test_selected = sfm.transform(X_test)

# Get the names of the selected features
feature_names = df.iloc[:, :-1].columns[sfm.get_support()]

# Fit the multinomial logistic regression model to the reduced feature set
mlr.fit(X_train_selected, y_train)

# Predict the test data
y_pred = mlr.predict(X_test_selected)

# Create a pandas dataframe to display the most important features and their coefficients
coef_df = pd.DataFrame({'feature': feature_names, 'coefficient': mlr.coef_[0]})
coef_df = coef_df.sort_values('coefficient', ascending=True)

# Print the dataframe
print(coef_df)
```

* The most important features are Current Liquidity, ROE, Cash, CFO, ST Debt, Sales/Revenues, ....

## 3. Model fitting and Evaluation

### 3.1. Multinomial logistic regression with feature selection

```python
df["Rating"].unique()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Calculating the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Calculating the accuracy score
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

### 3.2. Multinomial Regression with Principal Component Analysis

```python
from sklearn.decomposition import PCA

from sklearn.metrics import classification_report


# Performing PCA to reduce the dimensionality of the data
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Fiting the MLR model on the transformed data
mlr_pca = LogisticRegression(multi_class='multinomial', solver='lbfgs')
mlr_pca.fit(X_train_pca, y_train)

# Evaluating the performance of the model on the test set
y_pred = mlr_pca.predict(X_test_pca)
print(classification_report(y_test, y_pred))
```

### 3.3. SVC with Principal Component Analysis

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC

# Fiting the SVM model on the transformed data
svm = SVC()
svm.fit(X_train_pca, y_train)

# Evaluating the performance of the model on the test set
y_pred = svm.predict(X_test_pca)
print(classification_report(y_test, y_pred))
```

## 4. Hyperparameter tuning with Support Vector Machine model

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C': [0.1, 0.2, 0.3,1], 
              'gamma': [1, 0.1, 0.01],
              'kernel': ['rbf']}
svc = SVC()
grids = GridSearchCV(svc,parameters,cv=2)
svm_tuned = grids.fit(X_train_pca, np.ravel(y_train))
```

```python
print(svm_tuned.best_params_)
print(svm_tuned.best_score_)
```

## 5. Ensembling: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

estimator_list = np.arange(10,200,10)
for i in estimator_list:
    forest = RandomForestClassifier(n_estimators=i, random_state=1)
    forest.fit(X_train, y_train)
    
    # Making predictions on the training data
    y_pred = forest.predict(X_test)
    
    # Calculating the accuracy of the random forest classifier with train data
    accuracy = accuracy_score(y_test, y_pred)
    print(f"In Sample Accuracy with n_estimators {i}: {accuracy}")
```
